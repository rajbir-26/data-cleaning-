```python
import matplotlib.pyplot as plt
```

```python
top_categories = df['category'].value_counts().head(10)

plt.figure(figsize=(12,6))
top_categories.plot(kind='bar')

plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=90)

plt.show()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['discount_percentage'], df['rating'])

plt.title("Rating vs Discount Percentage")
plt.xlabel("Discount Percentage")
plt.ylabel("Rating")

plt.show()
```

```python
low = len(df[df['discount_percentage'] < 20])
medium = len(df[(df['discount_percentage'] >= 20) & (df['discount_percentage'] < 50)])
high = len(df[df['discount_percentage'] >= 50])

labels = ['Low Discount', 'Medium Discount', 'High Discount']
sizes = [low, medium, high]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Discount Distribution")
plt.show()
```

```python
top_rating = df['rating'].sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
plt.plot(top_rating)

plt.title("Top 10 Product Ratings")
plt.xlabel("Top Products")
plt.ylabel("Rating")

plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(df['rating'])

plt.title("Product Rating Box Plot")

plt.show()
```

```python
top20 = df.head(20)

plt.figure(figsize=(10,5))

plt.plot(top20['actual_price'], label='Actual Price')
plt.plot(top20['discounted_price'], label='Discounted Price')

plt.title("Actual Price vs Discounted Price")
plt.xlabel("Top 20 Products")
plt.ylabel("Price")

plt.legend()

plt.show()
```

```python
# Convert rating_count into numeric
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'])

# Top 10 products with highest rating count
top_products = df.nlargest(10, 'rating_count')

plt.figure(figsize=(10,6))

plt.barh(top_products['product_name'], top_products['rating_count'])

plt.title("Top 10 Products by Rating Count")
plt.xlabel("Rating Count")
plt.ylabel("Product Name")

plt.show()
```

```python
# Convert to numeric (safe even if already numeric)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Top 10 products by rating count
top_products = df.nlargest(10, 'rating_count')

plt.figure(figsize=(10,6))

plt.barh(top_products['product_name'], top_products['rating_count'])

plt.title("Top 10 Products by Rating Count")
plt.xlabel("Rating Count")
plt.ylabel("Product Name")

plt.show()
```
